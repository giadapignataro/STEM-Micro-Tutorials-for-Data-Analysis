## 1. Creating a basic line plot

Weâ€™ll use NumPy to generate simple data and visualize it with Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Create simple line plot
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()

#Want to save the plot? Try
#plt.savefig("plot.png")
```

> **Tip:** Use plt.plot() to create a line chart. Always include axis labels and titles to make your plots self-explanatory.

```python
# Adding style and color
plt.plot(x, y, color='skyblue', linestyle='--', marker='o', markersize=5)
plt.title("Nicer Line Plot")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True, linewidth=0.2)
plt.show()
```

>**Tip**: You can customize line color, style, and markers to make your plots more readable. Try changing color to 'coral' or 'navy' to experiment.

## 2. Plotting a histogram

```python
#Generate normally distributed data with NumPy 

data = np.random.normal(0, 1, 1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='grey', linewidth=0.5)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

>**Tip**: Histograms are perfect for showing the distribution of your data. Try to play with bin size, to understand how many bins are ideal to show the data.

## 3. Working with Colorbars
Colorbars help represent an additional variable, theyâ€™re essential for heatmaps or intensity data.

```python
#Generate a 10x10 matrix of random samples from a uniform distribution over [0, 1)

Z = np.random.rand(10,10)
plt.imshow(Z, cmap='viridis')
plt.title("Heatmap with Colorbar")
plt.colorbar(label='Intensity')
plt.show()
```

>**Tip**: you can find TONS of colormaps on Matplotlib documentation. Choose one that shows data in a non-confusing way, and that matches the presentation aesthetics.

## 4. Scatter Plots

Scatter plots show how two variables relate to each other. Each point represents an *observation*; for example, intensity vs time, or temperature vs energy.
You can also add color or size dimensions to encode even more information visually.

```python
# Generate sample data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
colors = x + y  # create color dimension
sizes = (x + y) * 500  # create size dimension

plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7, edgecolors='black')
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(label='[x + y]')
plt.show()
```

>**Tip**: Adding a colorbar gives your scatter plot a third dimension. It shows how combined values change across the dataset.
Adjusting *alpha* (transparency) helps avoid visual clutter when points overlap.
