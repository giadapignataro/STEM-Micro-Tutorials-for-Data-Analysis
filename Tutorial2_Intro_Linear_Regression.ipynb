## 1. Example Dataset 

We‚Äôll start with a simple dataset representing a linear relationship with some noise.

```python
import numpy as np
import matplotlib.pyplot as plt

#Generate simple data
np.random.seed(0)

#Linear relationship
x = np.linspace(0, 10, 20)
y = 2.5 * x + 1.0 + np.random.normal(0, 2, size=len(x))
```

Before fitting any model, always look at your data.

```python
#Create a simple plot to visualize the data

plt.scatter(x, y, color='skyblue', edgecolor='navy')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Observed data")
plt.grid(alpha=0.2)
plt.show()
```

> **Tip:** Even if it is just for you, a clear, well-organized plot is easier to read and interpret.

## 2. Fitting a Line to the Data

We now fit a straight line of the form:

y = m x + b

where:
- m is the slope
- b is the intercept


```python
#Use the function polyfit to get the coefficients from the fit

coefficients = np.polyfit(x, y, 1)
m, b = coefficients

print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
```

Now, let‚Äôs overlay the fitted line on the data.

```python
#Create the plot of the fitted line

y_fit = m * x + b

plt.scatter(x, y, color='skyblue', edgecolor='navy', zorder=10, label="Data")
plt.plot(x, y_fit, color="purple", label="Linear fit")
plt.xlabel("x")
plt.ylabel("y")
plt.legend() #add a legend to the plot
plt.title("Fit of observed data")
plt.grid(alpha=0.2)
plt.show()
```

> **Tip:** For better visualisation, you can decide which data is displayed on top with 'zorder'; also, remember to always add a legend when displaying multiple datasets.

## 3. Interpreting the Result

- The **slope** tells you how much y changes when x increases by one unit
- The **intercept** is the value of y when x = 0
